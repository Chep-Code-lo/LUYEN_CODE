## **üîπ M√É C++ (BIGINT)**  

```cpp
#include <iostream>
#include <vector>

using namespace std;

// L·ªõp BigInt h·ªó tr·ª£ ph√©p nh√¢n v√† gi·ªØ l·∫°i k ch·ªØ s·ªë cu·ªëi
struct BigInt {
    vector<int> digits; // L∆∞u tr·ªØ s·ªë theo t·ª´ng ch·ªØ s·ªë
    int mod; // Ch·ªâ gi·ªØ k ch·ªØ s·ªë cu·ªëi c√πng

    // Kh·ªüi t·∫°o BigInt t·ª´ s·ªë nguy√™n nh·ªè
    BigInt(int num, int k) {
        mod = 1;
        for (int i = 0; i < k; i++) mod *= 10; // 10^k
        while (num > 0) {
            digits.push_back(num % 10); // L∆∞u theo th·ª© t·ª± ng∆∞·ª£c
            num /= 10;
        }
    }

    // Nh√¢n BigInt v·ªõi s·ªë nguy√™n nh·ªè
    void multiply(int num) {
        int carry = 0;
        for (size_t i = 0; i < digits.size(); i++) {
            int prod = digits[i] * num + carry;
            digits[i] = prod % 10;
            carry = prod / 10;
        }
        while (carry > 0) {
            digits.push_back(carry % 10);
            carry /= 10;
        }

        // Ch·ªâ gi·ªØ l·∫°i k ch·ªØ s·ªë cu·ªëi c√πng
        while (digits.size() > 9) digits.pop_back();
    }

    // In s·ªë theo ƒë√∫ng th·ª© t·ª±
    void print(int k) {
        while (digits.size() < k) digits.push_back(0); // ƒê·∫£m b·∫£o c√≥ ƒë·ªß k ch·ªØ s·ªë
        for (int i = digits.size() - 1; i >= 0; i--) cout << digits[i];
        cout << endl;
    }
};

// H√†m t√≠nh m^n d√πng BigInt
void powerBigInt(int m, int n, int k) {
    BigInt result(1, k);
    for (int i = 0; i < n; i++) result.multiply(m);
    result.print(k);
}

int main() {
    int k, m, n;

    // Nh·∫≠p d·ªØ li·ªáu
    cout << "Nh·∫≠p k, m, n: ";
    cin >> k >> m >> n;

    // T√¨m k ch·ªØ s·ªë cu·ªëi c·ªßa m^n
    cout << "K·∫øt qu·∫£: ";
    powerBigInt(m, n, k);

    return 0;
}
```

**Coder:**

```cpp
/***************************************
 * **T√åM USCLN (GCD) & BSCNN (LCM) B·∫∞NG BIGINT**
 * - T√¨m **∆Ø·ªõc s·ªë chung l·ªõn nh·∫•t (USCLN - GCD)**
 * - T√¨m **B·ªôi s·ªë chung nh·ªè nh·∫•t (BSCNN - LCM)**
 * - X·ª≠ l√Ω s·ªë l·ªõn b·∫±ng **BigInt**
 ***************************************/

#include <iostream>
#include <vector>
#include <algorithm>

using namespace std;

/**
 * **L·ªöP BIGINT**: H·ªó tr·ª£ x·ª≠ l√Ω s·ªë l·ªõn b·∫±ng chu·ªói
 */
struct BigInt {
    string value; // Chu·ªói l∆∞u s·ªë nguy√™n l·ªõn

    // **Constructor**: Kh·ªüi t·∫°o t·ª´ string ho·∫∑c s·ªë nguy√™n
    BigInt(string val) { value = removeLeadingZeros(val); }
    BigInt(long long num) { value = to_string(num); }

    // **H√†m x√≥a s·ªë 0 ·ªü ƒë·∫ßu ƒë·ªÉ tr√°nh l·ªói**
    static string removeLeadingZeros(string str) {
        while (str.size() > 1 && str[0] == '0') str.erase(0, 1);
        return str;
    }

    // **To√°n t·ª≠ so s√°nh (==, <)**
    bool operator==(const BigInt &other) const { return value == other.value; }
    bool operator<(const BigInt &other) const {
        if (value.size() != other.value.size()) return value.size() < other.value.size();
        return value < other.value;
    }

    // **To√°n t·ª≠ nh√¢n BigInt v·ªõi s·ªë nguy√™n nh·ªè (int)**
    BigInt operator*(int num) const {
        int carry = 0;
        string result = "";
        for (int i = value.size() - 1; i >= 0; i--) {
            int prod = (value[i] - '0') * num + carry;
            result = char(prod % 10 + '0') + result;
            carry = prod / 10;
        }
        if (carry) result = char(carry + '0') + result;
        return BigInt(result);
    }

    // **To√°n t·ª≠ chia l·∫•y ph·∫ßn nguy√™n**
    BigInt operator/(int num) const {
        string result = "";
        int rem = 0;
        for (char c : value) {
            rem = rem * 10 + (c - '0');
            result += char(rem / num + '0');
            rem %= num;
        }
        return BigInt(removeLeadingZeros(result));
    }

    // **H√†m tr·ª´ hai BigInt**
    BigInt operator-(const BigInt &other) const {
        string a = value, b = other.value;
        if (a == b) return BigInt("0");
        if (a < b) swap(a, b); // ƒê·∫£m b·∫£o a >= b

        string result = "";
        int carry = 0;
        for (int i = 0; i < a.size(); i++) {
            int x = (i < a.size() ? a[a.size() - 1 - i] - '0' : 0);
            int y = (i < b.size() ? b[b.size() - 1 - i] - '0' : 0) + carry;
            if (x < y) {
                x += 10;
                carry = 1;
            } else {
                carry = 0;
            }
            result = char((x - y) + '0') + result;
        }
        return BigInt(removeLeadingZeros(result));
    }
};

/**
 * **H√†m t√≠nh USCLN (GCD) c·ªßa hai s·ªë BigInt**
 */
BigInt gcd(BigInt a, BigInt b) {
    while (!(b == BigInt("0"))) {
        BigInt temp = b;
        b = a - (a / stoi(b.value)) * stoi(b.value);
        a = temp;
    }
    return a;
}

/**
 * **H√†m t√≠nh BSCNN (LCM) c·ªßa hai s·ªë BigInt**
 */
BigInt lcm(BigInt a, BigInt b) {
    return (a * stoi(b.value)) / stoi(gcd(a, b).value);
}

/**
 * **H√†m t√¨m USCLN v√† BSCNN c·ªßa n s·ªë trong m·∫£ng**
 */
pair<BigInt, BigInt> findGCD_LCM(vector<BigInt> &arr) {
    BigInt GCD = arr[0];
    BigInt LCM = arr[0];

    for (int i = 1; i < arr.size(); i++) {
        GCD = gcd(GCD, arr[i]);
        LCM = lcm(LCM, arr[i]);
    }
    return {GCD, LCM};
}

/**
 * **H√†m main: Nh·∫≠p v√† x·ª≠ l√Ω d·ªØ li·ªáu**
 */
int main() {
    int n;
    cout << "Nh·∫≠p s·ªë l∆∞·ª£ng ph·∫ßn t·ª≠ n (n ‚â§ 10): ";
    cin >> n;

    vector<BigInt> arr;
    cout << "Nh·∫≠p " << n << " s·ªë nguy√™n l·ªõn:\n";
    for (int i = 0; i < n; i++) {
        string num;
        cin >> num;
        arr.push_back(BigInt(num));
    }

    auto [GCD, LCM] = findGCD_LCM(arr);
    
    cout << "USCLN (GCD): " << GCD.value << endl;
    cout << "BSCNN (LCM): " << LCM.value << endl;

    return 0;
}
```

D∆∞·ªõi ƒë√¢y l√† code C++ s·ª≠ d·ª•ng th∆∞ vi·ªán `BigInteger` t·ª´ **Boost** ho·∫∑c c√≥ th·ªÉ d√πng `string` ƒë·ªÉ x·ª≠ l√Ω s·ªë l·ªõn:

```cpp
#include <bits/stdc++.h>
using namespace std;

// H√†m c·ªông hai s·ªë l·ªõn d∆∞·ªõi d·∫°ng chu·ªói
string addBigNum(string a, string b) {
    string res = "";  // Chu·ªói k·∫øt qu·∫£
    int carry = 0, sum;
    int n = a.size(), m = b.size();

    // ƒê·∫£m b·∫£o a l√† s·ªë c√≥ nhi·ªÅu ch·ªØ s·ªë h∆°n ho·∫∑c b·∫±ng b
    if (n < m) swap(a, b), swap(n, m);

    // ƒê·∫£o ng∆∞·ª£c chu·ªói ƒë·ªÉ t√≠nh to√°n t·ª´ h√†ng ƒë∆°n v·ªã tr∆∞·ªõc
    reverse(a.begin(), a.end());
    reverse(b.begin(), b.end());

    // C·ªông t·ª´ng ch·ªØ s·ªë
    for (int i = 0; i < n; i++) {
        sum = (i < m ? (b[i] - '0') : 0) + (a[i] - '0') + carry;
        carry = sum / 10; // L∆∞u gi√° tr·ªã nh·ªõ
        res += (sum % 10) + '0'; // L·∫•y ph·∫ßn ƒë∆°n v·ªã
    }

    // N·∫øu c√≤n s·ªë nh·ªõ, th√™m v√†o k·∫øt qu·∫£
    if (carry) res += (carry + '0');

    // ƒê·∫£o ng∆∞·ª£c chu·ªói l·∫°i ƒë·ªÉ c√≥ k·∫øt qu·∫£ ƒë√∫ng
    reverse(res.begin(), res.end());
    return res;
}

// H√†m so s√°nh hai s·ªë l·ªõn d∆∞·ªõi d·∫°ng chu·ªói
bool isGreaterOrEqual(string a, string b) {
    if (a.size() != b.size()) return a.size() > b.size();
    return a >= b;
}

// H√†m ƒë·∫øm s·ªë Fibonacci trong ƒëo·∫°n [a, b]
int countFibonacciInRange(string a, string b) {
    string f0 = "0", f1 = "1", fn;
    int count = 0;

    // Sinh d√£y Fibonacci v√† ki·ªÉm tra s·ªë n·∫±m trong ƒëo·∫°n [a, b]
    while (isGreaterOrEqual(b, f1)) {  // D·ª´ng n·∫øu f1 > b
        if (isGreaterOrEqual(f1, a))   // Ki·ªÉm tra n·∫øu f1 >= a
            count++;
        fn = addBigNum(f0, f1); // T√≠nh s·ªë Fibonacci ti·∫øp theo
        f0 = f1;                // C·∫≠p nh·∫≠t f(n-2)
        f1 = fn;                // C·∫≠p nh·∫≠t f(n-1)
    }

    return count; // Tr·∫£ v·ªÅ t·ªïng s·ªë Fibonacci trong ƒëo·∫°n [a, b]
}

int main() {
    // Kh·ªüi t·∫°o gi√° tr·ªã a = 10^199, b = 10^200
    string a = "1" + string(199, '0');  // 10^199
    string b = "10" + string(199, '0'); // 10^200

    // G·ªçi h√†m v√† in k·∫øt qu·∫£
    cout << countFibonacciInRange(a, b) << endl;
    return 0;
}
```

```cpp
#include <bits/stdc++.h>
using namespace std;
using ll = long long;

// H√†m s√†ng Eratosthenes ƒë·ªÉ t√¨m c√°c s·ªë nguy√™n t·ªë ‚â§ n
vector<int> sieve(int n) {
    vector<bool> isPrime(n + 1, true);
    vector<int> primes;

    for (int i = 2; i <= n; i++) {
        if (isPrime[i]) {
            primes.push_back(i);
            for (int j = i * 2; j <= n; j += i)
                isPrime[j] = false;
        }
    }
    return primes;
}

// H√†m t√≠nh s·ªë l·∫ßn xu·∫•t hi·ªán c·ªßa s·ªë nguy√™n t·ªë p trong n!
ll countExponent(int n, int p) {
    ll exp = 0;
    for (ll power = p; power <= n; power *= p) {
        exp += n / power;
    }
    return exp;
}

// H√†m t√≠nh s·ªë ∆∞·ªõc c·ªßa n!
ll countDivisors(int n) {
    vector<int> primes = sieve(n);
    ll divisors = 1;

    for (int p : primes) {
        ll exp = countExponent(n, p);
        divisors *= (exp + 1); // C√¥ng th·ª©c d(n!) = (e_2 + 1) * (e_3 + 1) * ...
    }
    return divisors;
}

// H√†m nh√¢n hai s·ªë l·ªõn (d√πng chu·ªói)
string multiplyBigNum(string a, string b) {
    int n = a.size(), m = b.size();
    vector<int> result(n + m, 0);

    for (int i = n - 1; i >= 0; i--) {
        for (int j = m - 1; j >= 0; j--) {
            int mul = (a[i] - '0') * (b[j] - '0');
            int sum = mul + result[i + j + 1];

            result[i + j] += sum / 10;
            result[i + j + 1] = sum % 10;
        }
    }

    string res = "";
    for (int num : result) {
        if (!(res.empty() && num == 0)) // Lo·∫°i b·ªè s·ªë 0 th·ª´a
            res.push_back(num + '0');
    }
    return res.empty() ? "0" : res;
}

// H√†m t√≠nh t·ªïng c√°c ∆∞·ªõc c·ªßa n! (BigNum)
string sumDivisors(int n) {
    vector<int> primes = sieve(n);
    string sumDiv = "1";

    for (int p : primes) {
        ll exp = countExponent(n, p);
        string p_pow = "1";
        string p_str = to_string(p);

        // T√≠nh p^(e_p+1) b·∫±ng nh√¢n chu·ªói
        for (int i = 0; i <= exp; i++)
            p_pow = multiplyBigNum(p_pow, p_str);

        // T√≠nh (p^(e_p+1) - 1) / (p - 1)
        string numerator = multiplyBigNum(p_pow, "1");
        numerator[numerator.size() - 1] -= 1; // Gi·∫£m ƒëi 1

        string denominator = to_string(p - 1);

        // Nh√¢n v√†o t·ªïng c√°c ∆∞·ªõc
        sumDiv = multiplyBigNum(sumDiv, numerator);
    }

    return sumDiv;
}

int main() {
    int n;
    cout << "Nh·∫≠p n: ";
    cin >> n;

    // T√≠nh s·ªë ∆∞·ªõc c·ªßa n!
    ll numDivisors = countDivisors(n);
    cout << "S·ªë ∆∞·ªõc c·ªßa " << n << "! l√†: " << numDivisors << endl;

    // T√≠nh t·ªïng c√°c ∆∞·ªõc c·ªßa n!
    cout << "T·ªïng c√°c ∆∞·ªõc c·ªßa " << n << "! l√†: " << sumDivisors(n) << endl;

    return 0;
}
```

### üîπ **4. C++ Implementation v·ªõi BigNum (T·ª± X√¢y)**
```cpp
#include <iostream>
#include <vector>
using namespace std;

// Struct BigNum ƒë·ªÉ bi·ªÉu di·ªÖn s·ªë l·ªõn
struct BigNum {
    vector<int> digits;

    // H√†m kh·ªüi t·∫°o t·ª´ s·ªë nguy√™n
    BigNum(long long num = 0) {
        while (num > 0) {
            digits.push_back(num % 10);
            num /= 10;
        }
        if (digits.empty()) digits.push_back(0); // X·ª≠ l√Ω s·ªë 0
    }

    // H√†m chia BigNum cho s·ªë nguy√™n nh·ªè (int)
    BigNum operator/(int divisor) const {
        BigNum result;
        result.digits.clear();
        long long carry = 0;
        for (int i = digits.size() - 1; i >= 0; --i) {
            carry = carry * 10 + digits[i];
            result.digits.push_back(carry / divisor);
            carry %= divisor;
        }
        reverse(result.digits.begin(), result.digits.end());
        while (result.digits.size() > 1 && result.digits.back() == 0) {
            result.digits.pop_back(); // Lo·∫°i b·ªè s·ªë 0 ·ªü ƒë·∫ßu
        }
        return result;
    }

    // H√†m c·ªông BigNum v·ªõi s·ªë nguy√™n nh·ªè (int)
    BigNum operator+(int num) const {
        BigNum result = *this;
        int carry = num, i = 0;
        while (carry > 0 || i < result.digits.size()) {
            if (i == result.digits.size()) result.digits.push_back(0);
            carry += result.digits[i];
            result.digits[i] = carry % 10;
            carry /= 10;
            i++;
        }
        return result;
    }

    // H√†m ki·ªÉm tra BigNum c√≥ l·ªõn h∆°n s·ªë nguy√™n kh√¥ng
    bool operator<=(const BigNum &other) const {
        if (digits.size() != other.digits.size()) {
            return digits.size() < other.digits.size();
        }
        for (int i = digits.size() - 1; i >= 0; --i) {
            if (digits[i] != other.digits[i]) return digits[i] < other.digits[i];
        }
        return true;
    }
};

// H√†m ƒë·∫øm s·ªë ch·ªØ s·ªë 0 t·∫≠n c√πng c·ªßa n!
BigNum countTrailingZeros(BigNum n) {
    BigNum count(0);
    BigNum five(5);
    while (five <= n) {
        count = count + (n / 5); // count += n / 5
        five = five + five * 4;  // five *= 5
    }
    return count;
}

int main() {
    string str;
    cout << "Nh·∫≠p n: ";
    cin >> str;

    // Chuy·ªÉn chu·ªói th√†nh BigNum
    BigNum n;
    for (int i = str.size() - 1; i >= 0; --i) {
        n.digits.push_back(str[i] - '0');
    }

    // T√≠nh s·ªë ch·ªØ s·ªë 0 t·∫≠n c√πng c·ªßa n!
    BigNum result = countTrailingZeros(n);

    // Xu·∫•t k·∫øt qu·∫£
    cout << "S·ªë ch·ªØ s·ªë 0 t·∫≠n c√πng c·ªßa " << str << "! l√†: ";
    for (int i = result.digits.size() - 1; i >= 0; --i) {
        cout << result.digits[i];
    }
    cout << endl;
    return 0;
}
```

#**2.10 T√≠nh $( C_n^k $)**

**Coder:**

$( C_n^k $) ƒë∆∞·ª£c ƒë·ªãnh nghƒ©a b·ªüi c√¥ng th·ª©c : $C_n^k = \frac{n!}{k!(n-k)!}$
r√†ng bu·ªôc: $(0 < k \leq n \leq 2000)$ h√£y t√≠nh $(c_k^n$) v·ªõi x·ª≠ l√Ω bignum t·ª± x√¢y


**ChatGPT:**

## üìå **T√≠nh t·ªï h·ª£p $( C_n^k $) (T·ªï h·ª£p ch·∫≠p \( k \) c·ªßa \( n \))**

### üîπ **1. ƒê·ªãnh nghƒ©a t·ªï h·ª£p**
- C√¥ng th·ª©c t√≠nh t·ªï h·ª£p:
  $C_n^k = \frac{n!}{k!(n-k)!}$
- V√≠ d·ª•:
  $$
  C_3^5 = \frac{5!}{3!(5-3)!} = \frac{5 \times 4 \times 3!}{3! \times 2!} = \frac{5 \times 4}{2!} = 10
  $$

---

### üîπ **2. C√°ch gi·∫£i v·ªõi BigNum**
- Tr√°nh t√≠nh giai th·ª´a l·ªõn tr·ª±c ti·∫øp v√¨ d·ªÖ g√¢y tr√†n s·ªë.
- C√¥ng th·ª©c t·ªëi ∆∞u:
  $$
  C_n^k = \frac{n(n-1)(n-2)...(n-k+1)}{k!}
  $$
- L∆∞u √Ω: $( C_n^k = C_n^{n-k} $), gi·∫£m s·ªë ph√©p to√°n.

### üîπ **3. C++ Implementation v·ªõi BigNum**
```cpp
#include <iostream>
#include <vector>
using namespace std;

// Struct BigNum ƒë·ªÉ x·ª≠ l√Ω s·ªë l·ªõn
struct BigNum {
    vector<int> digits;

    // Kh·ªüi t·∫°o t·ª´ s·ªë nguy√™n
    BigNum(long long num = 0) {
        while (num > 0) {
            digits.push_back(num % 10);
            num /= 10;
        }
        if (digits.empty()) digits.push_back(0);
    }

    // Nh√¢n BigNum v·ªõi s·ªë nguy√™n nh·ªè
    BigNum operator*(int num) const {
        BigNum result;
        result.digits.clear();
        long long carry = 0;
        for (int d : digits) {
            carry += (long long)d * num;
            result.digits.push_back(carry % 10);
            carry /= 10;
        }
        while (carry > 0) {
            result.digits.push_back(carry % 10);
            carry /= 10;
        }
        return result;
    }

    // Chia BigNum cho s·ªë nguy√™n nh·ªè
    BigNum operator/(int num) const {
        BigNum result;
        result.digits.clear();
        long long carry = 0;
        for (int i = digits.size() - 1; i >= 0; --i) {
            carry = carry * 10 + digits[i];
            result.digits.push_back(carry / num);
            carry %= num;
        }
        reverse(result.digits.begin(), result.digits.end());
        while (result.digits.size() > 1 && result.digits.back() == 0) result.digits.pop_back();
        return result;
    }

    // In BigNum
    void print() const {
        for (int i = digits.size() - 1; i >= 0; --i) cout << digits[i];
        cout << endl;
    }
};

// H√†m t√≠nh t·ªï h·ª£p C_k^n v·ªõi BigNum
BigNum combination(int n, int k) {
    if (k > n - k) k = n - k; // T·∫≠n d·ª•ng t√≠nh ƒë·ªëi x·ª©ng

    BigNum num(1), denom(1);
    for (int i = 0; i < k; i++) {
        num = num * (n - i); // T·ª≠ s·ªë: n * (n-1) * ... * (n-k+1)
        denom = denom * (i + 1); // M·∫´u s·ªë: k!
    }
    return num / denom; // Chia t·ª≠ cho m·∫´u
}

int main() {
    int n, k;
    cout << "Nh·∫≠p n, k: ";
    cin >> n >> k;

    cout << "C_" << k << "^" << n << " = ";
    combination(n, k).print();

    return 0;
}
```
